```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def ReadJson(InFile):
    with open(InFile, 'rb') as f:
        data = f.readlines()

    # remove the trailing "\n" from each line
    data = map(lambda x: x.rstrip(), data)

    # each element of 'data' is an individual JSON object.
    # i want to convert it into an *array* of JSON objects
    # which, in and of itself, is one large JSON object
    # basically... add square brackets to the beginning
    # and end, and have all the individual business JSON objects
    # separated by a comma
    data_json_str = "[" + ','.join(data) + "]"

    # now, load it into pandas
    return pd.read_json(data_json_str)

dfTestBusiness = ReadJson('D:\Yelp2013\yelp_test_set\yelp_test_set_business.json')
dfTestCheckin = ReadJson('D:\Yelp2013\yelp_test_set\yelp_test_set_checkin.json')
dfTestReview = ReadJson('D:\Yelp2013\yelp_test_set\yelp_test_set_review.json')
dfTestUser = ReadJson('D:\Yelp2013\yelp_test_set\yelp_test_set_user.json')

dfTrainBusiness = ReadJson('D:\Yelp2013\yelp_training_set\yelp_training_set_business.json')
dfTrainCheckin = ReadJson('D:\Yelp2013\yelp_training_set\yelp_training_set_checkin.json')
dfTrainReview = ReadJson('D:\Yelp2013\yelp_training_set\yelp_training_set_review.json')
dfTrainUser = ReadJson('D:\Yelp2013\yelp_training_set\yelp_training_set_user.json')
```
